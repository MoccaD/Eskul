## Tipe data koleksi -> List, Set, Tuple, Dict

```python
# List
List_buah = ["Apel", "Mangga", "Pisang", "Mangga"]
print(List_buah)
print(type(List_buah))
print(List_buah[1])

List_buah[1] = "Durian"
print(List_buah)
```

```python
# Tuple
List_buah = ("Apel", "Mangga", "Pisang", "Mangga")
print(List_buah)
print(type(List_buah))
print(List_buah[1])

# List_buah[1] = "Durian" -> Tidak bisa
# print(List_buah)
```

```python
# Set
List_buah = {"Apel", "Mangga", "Pisang", "Mangga"}
print(List_buah)
print(type(List_buah))
# print(List_buah[1]) -> Tidak bisa

List_buah.add("Jeruk")
print(List_buah)
List_buah.remove("Pisang")
print(List_buah)
```

```python
# Dict
Nama_mobil = {
    "mobil 1": {
        "nama":"Ford",
        "tahun":1964
        },
    "mobil 1": {
        "nama":"Ford",
        "tahun":1945
        }
      }

print(Nama_mobil)
print(type(Nama_mobil))
# print(Nama_mobil[0])

Nama_mobil["mobil 1"]["tahun"] = 2000
print(Nama_mobil)
```

# Operator

```python
angka_1 = 14
angka_2 = 3

# Penjumlahan
print("Penjumlahan : " , angka_1 + angka_2)

# Pengurangan
print("Pengurangan : ", angka_1 - angka_2)

# Perkalian
print("Perkalian : ", angka_1 * angka_2)

# Pembagian
print("Pembagian : ", angka_1 / angka_2)

# Pangkat
print("Pangkat : ", angka_1 ** angka_2)

# Modulo
print("Modulo : ", angka_1 % angka_2)

# Division
print("Division : ", angka_1 // angka_2)
```

```python
def luas_lingkaran(r: int=1):
  """Menghitung luas lingkaran dengan rumus A = π * r^2"""
  area = 3.14 * (r ** 2)
  return area

print(f"area: {luas_lingkaran():.2f}")
print(f"area: {luas_lingkaran(8):.2f}")
```
